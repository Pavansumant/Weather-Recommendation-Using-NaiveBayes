```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
import numpy as np
```

```python
df=pd.read_csv("/kaggle/input/tennis/tennis.csv")
df
```

```python
df.isna().sum()
```

```python
inputs=df.drop('play',axis=1)
target=df["play"]
```

```python
inputs
```

```python
target
```

```python
l=LabelEncoder()
```

```python
inputs["outlook_n"]=l.fit_transform(df["outlook"])
inputs["temp_n"]=l.fit_transform(df["temp"])
inputs["humidity_n"]=l.fit_transform(df["humidity"])
inputs["windy_n"]=l.fit_transform(df["windy"])
```

```python
inputs
```

```python
classifier=GaussianNB()
classifier.fit(inputs,target)
```

```python
classifier.score(inputs,target)
```

```python
classifier.predict(np.array([2,1,0,1]).reshape(1,-1))
```
